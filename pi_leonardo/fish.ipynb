### Questão 1 (1,0 pt)

Descreva os conceitos de _precision_ e _recall_ com suas palavras. Construa um exemplo de problema de classificação - que não seja nem médico, nem jurídico, para não reutilizar os exemplos de sala de aula - e discuta neste exemplo as consequências de

- um alto _precision_ e baixo _recall_
- um alto _recall_ e um baixo _precision_.


**Resposta:**

Tanto o **precision** quanto o **recall** são métricas de avaliação de modelos de classificação. O **precision** é a proporção de elementos classificadas como positivos que realmente são positivos, definido pela seguinte fórmula:

$$\text{precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}$$

Já o **recall** é a proporção de elementos positivos que foram classificados como positivos. É definido pela seguinte fórmula:

$$\text{recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}$$

Onde $TP$, $FP$ e $FN$ são, respectivamente, o número de verdadeiros positivos, falsos positivos e falsos negativos. Ao analisar as fórmulas, percebe-se que eles são métricas complementares, ou seja, se um aumenta, o outro diminui, já que sempre que o FP aumenta, o FN diminui, e vice-versa. Em outras palavras, se o modelo tem um alto precision, ele tem um baixo recall, e o contrário também é verdadeiro.

Assim, o **precision** é uma métrica que diz o quão preciso é o modelo em classificar exemplos positivos, ou seja, mede a qualidade das previsões positivas. Já o **recall** é uma métrica que diz o quão completo é o modelo em classificar exemplos positivos, ou seja, mede a capacidade do modelo em identificar corretamente todos os exemplos positivos no conjunto de dados.

Uma situação que é interessante ter um modelo com alto precision e baixo recall é na identificação de armas e drogas em malas em Aeroportos, uma vez que essas checagens acontecem milhares de vezes ao dia, e um modelo com maior recall, faria com que a polícia tivesse que parar para checar malas com uma frequência muito maior, o que seria inviável, e poderia gerar um grande problema logístico. Nesse caso, um precision mais alto, ajudará a garantir que apenas as malas que realmente contenham armas ou drogas sejam paradas para checagem, o que é muito mais eficiente.

Já uma situação que é interessante ter um modelo com alto recall e baixo precision é na identificação de problemas e falhas em ‘softwares’. Nessa situação é interessante abranger o maior número de problemas possíveis, mesmo que signifique que haja alguns falsos positivos. Um modelo com alto precision, poderia acabar não identificando alguns problemas que poderiam ser críticos para o ‘software’, o que pode ser extremamente prejudicial. Nesse caso, um recall mais alto ajudará a garantir que todas as possíveis vulnerabilidades de segurança sejam identificadas.


### Questão 2: (1,0 pt)

Em um problema de _machine learning_ temos uma _feature_ discreta com os seguintes valores possíveis: "Aprendiz", "Junior", "Senior", "Especialista". Estas categorias estão em ordem: um "Aprendiz" evolui para um "Junior", etc. Como esta _feature_ deve ser representada em um projeto de regressão linear? Explique vantagens e desvantagens da representação que você escolheu. Dica: não tem só uma resposta certa aqui, mas não preciso de todas as respostas certas possíveis não, apenas de uma que você escolher.


**Resposta:**

Para apropriar uma feature categórica ordinal, ou seja, uma feature que possui um conjunto finito e discreto de valores, e esses valores possuem uma ordem, podemos utilizar a técnica de **classificação ordinal**. Nessa técnica, cada valor da feature é mapeado para um número inteiro, e esse número é utilizado para treinar o modelo, de modo que preserve a ordem da categoria. Por exemplo, para essa feature com valores "Aprendiz", "Júnior", "Sênior" e "Especialista", podemos mapear “Aprendiz” para 0, “Júnior” para 1, “Sênior” para 2 e "Especialista" para 3. Assim, os valores foram mapeados para números inteiros, fazendo com que a feature se torne numérica, permitindo que ela seja utilizada em modelos de regressão.

A principal vantagem dessa abordagem é a manutenção da ordem das categorias, o que pode ser interessante quando essa ordem é relevante para o problema. Por exemplo, se o problema for prever o salário de um profissional, a ordem é muito importante ser mantida, já que o salário de um aprendiz é comumente menor do que o de um júnior, e assim por diante. Uma desvantagem da utilização da classificação ordinal, é que ela não considera a distância entre os valores, tornando alguns modelos de machine learning, que usam medidas de distância, como a regressão linear, imprecisos.

### Questão 4: (1,0 pt)

Suponha que você está explicando _machine learning_ _para seus colegas do primeiro ano, fazendo ciência dos dados_. Explique como funciona o _gradient descent_ na busca de melhores parâmetros para um modelo linear.


**Resposta:**

O gradient descent é um algoritmo de otimização que é utilizado para encontrar o mínimo de uma função de custo, que é calculada por uma série de parâmetros. Para fazer isso, o algorítimo vai ajustando esses parâmetros, até que a função de custo seja minimizada. Esse algorítimo é utilizado na regressão linear porque ela é uma técnica que depende da minimização da função de custo.

O objetivo é ajustar uma linha reta aos dados observados para prever a relação entre duas variáveis, logo, os parâmetros a serem ajustados são `a` ( coeficiente angular) e `b` (coeficiente linear). Para encontrar os valores de `a` e `b`, acontece o seguinte processo:

- Comecar com valores aleatórios para `a` e `b`
- Calcular o erro para cada ponto
- Calcular o erro total
- Calcular o gradiente da função de custo
- Atualizar os valores de `a` e `b` na direção oposta ao gradiente
- Repetir até que o erro seja pequeno o suficiente (um threshold pré definido)

Assim, a função da reta é definida por `y = a*x + b`, e é possível prever o valor da variável dependente `y` a partir do valor da variável independente `x`.

## Tratamento de dados


Remover outiliers para deixar a análise mais precisa. Como são poucos dados, podemos remover os outliers sem perder muita informação.


```python
data_transformed = data.copy()

# Remover outliers
for col in data_transformed.select_dtypes("number").columns:
    q1 = data_transformed[col].quantile(0.25)
    q3 = data_transformed[col].quantile(0.75)
    iqr = q3 - q1
    data_transformed = data_transformed[
        (data_transformed[col] >= q1 - 1.5 * iqr)
        & (data_transformed[col] <= q3 + 1.5 * iqr)
    ]
```

Estimar o peso por meio de outros parâmetros é um problema de `regressão`. Sendo assim, as variáveis independentes devem ser numéricas. Uma forma de transformar a variável `Species` que é categorica, em numérica, é utilizando a técnica de `one-hot encoding`. Entretanto, como analisado na análise exploratória, não há uma espécie que se destaca das outras, para determinar o peso. Sendo assim, não há necessidade de transformar a variável `Species` em numérica. Assim, a variável `Species` será excluída do dataset.


```python
data_transformed.drop("Species", axis=1, inplace=True)
```

Uma outra constatação feita na análise exploratória, é que as variáveis `Length1`, `Length2`, e `Length3` possuem uma correlação muito alta entre si. Sendo assim, para evitar problemas de multicolinearidade, podemos criar uma nova variável chamada `BodyLength`, que será a multiplicação das variáveis `Length1`, `Length2`, e `Length3`. Assim, as variáveis `Length1`, `Length2`, e `Length3` serão excluídas do dataset.


```python
data_transformed["BodyLength"] = data_transformed.apply(
    lambda x: x["Length1"] + x["Length2"] + x["Length3"], axis=1
)
data_transformed.drop(["Length1", "Length2", "Length3"], axis=1, inplace=True)
```

## Aplicação do primeiro modelo


### Separando os dados em treino e teste


```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
X = data_transformed.drop("Weight", axis=1)
y = data_transformed["Weight"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)
```

### Aplicando o modelo


Podemos utilizar o Random Forest Regressor, que é um modelo de regressão baseado em árvores de decisão.


```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, random_state=RANDOM_STATE)
regressor.fit(X_train, y_train)
```

### Avaliando o primeiro modelo

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


def get_model_evaluation(y_test, y_pred):
    """Print model evaluation metrics"""
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    return mse, rmse, mae, r2


def print_evaluation(y_test, y_pred):
    """Print model evaluation metrics"""
    mse, rmse, mae, r2 = get_model_evaluation(y_test, y_pred)
    print(f"Mean Squared Error: {mse}")
    print(f"Root Mean Squared Error: {rmse}")
    print(f"Mean Absolute Error: {mae}")
    print(f"R2 Score: {r2}")


def plot_actual_vs_predicted(y_test, y_pred):
    """Plot the prediction"""
    plt.scatter(list(range(len(y_test))), y_test, label="Actual", color="red")
    plt.scatter(list(range(len(y_pred))), y_pred, label="Predicted", color="blue")

    plt.xlabel("Index")
    plt.ylabel("Weight")
    plt.legend()
    plt.show()


def plot_feature_importance(model, X):
    """Plot the feature importance"""
    n_features = X.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align="center")
    plt.yticks(np.arange(n_features), X.columns)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.show()
```

```python
# Avaliação do modelo básica
y_pred = regressor.predict(X_test)

print_evaluation(y_test, y_pred)
plot_actual_vs_predicted(y_test, y_pred)
```

## Utilizando outros modelos

Obtemos a performance do modelo utilizando o `Random Forest Regressor`, mas é importante avaliar se outros algorítimos de regressão podem ser mais eficientes. Assim, vamos aplicar outros modelos de regressão e comparar os resultados.


```python
# Tryinng to use other regression models and compare the results
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

models = {
    "Linear": LinearRegression(),
    "DecisionTree": DecisionTreeRegressor(),
    "KNeighbors": KNeighborsRegressor(),
    "SVR": SVR(),
    "RandomForest": RandomForestRegressor(),
}

# Plot all errors for each model in a single plot

mses = []
rmses = []
maes = []
r2s = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse, rmse, mae, r2 = get_model_evaluation(y_test, y_pred)
    mses.append(mse)
    rmses.append(rmse)
    maes.append(mae)
    r2s.append(r2)


def plot_comparison(labels, values, title):
    """Plot the comparison between models"""
    plt.bar(labels, values)
    plt.title(title)
    plt.show()


plot_comparison(list(models.keys()), mses, "Mean Squared Error")
plot_comparison(list(models.keys()), rmses, "Root Mean Squared Error")
plot_comparison(list(models.keys()), maes, "Mean Absolute Error")
plot_comparison(list(models.keys()), r2s, "R2 Score")
```

## Melhorando o modelo de Random Forest Regressor

Comparando os algorítmos, é possível observar que o `Random Forest Regressor` foi o que obteve o melhor resultado, em todas as métricas, sendo elas:

- `Mean Absolute Error` (MAE): Valor médio do erro absoluto, mensurando a diferença entre o valor real e o valor predito.
- `Mean Squared Error` (MSE): Valor médio do erro quadrático, mensurando a diferença entre o valor real e o valor predito, porém com maior peso para erros maiores, já que o ele é elevado ao quadrado.
- `Root Mean Squared Error` (RMSE): Valor médio da raiz dos erros quadráticos. É uma medida de erro que é mais fácil de interpretar, já que é expressa na mesma unidade da variável dependente.
- `R2 Score`: Valor que varia entre 0 e 1, indicando o quanto o modelo consegue explicar a variância dos dados.

Sendo assim, vamos melhorar o modelo de `Random Forest Regressor`.

```python
# Plot the feature importance
plot_feature_importance(regressor, X_train)
```

Pode-se obervar que a variável `Height` não é muito importante para determinar o `Weight`, no melhor modelo. Sendo assim, podemos remover essa variável do dataset e treinar novamente o modelo.

```python
new_data = data_transformed.copy()

# Remove Height
new_data.drop("Height", axis=1, inplace=True)

X = new_data.drop("Weight", axis=1)
y = new_data["Weight"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

regressor = RandomForestRegressor(n_estimators=100, random_state=RANDOM_STATE)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print_evaluation(y_test, y_pred)
plot_actual_vs_predicted(y_test, y_pred)
```
